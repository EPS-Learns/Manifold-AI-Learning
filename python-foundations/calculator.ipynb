```python
class circle():
    def __init__(self, u_radius):
        self.radius = u_radius

    def area(self):
        return 3.14 * self.radius ** 2

    def circumference(self):
        return 2 * 3.14 * self.radius
    
```

```python
c = circle(5)
print(c.radius)
print(c.area())
print(c.circumference())
```
